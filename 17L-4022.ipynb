```python
import spacy
import re
import string
import math
import random

sp = spacy.blank("en")
```

```python
def unigramFreq(corpus):
  uniFreq = {}
  count = 0
  for w in corpus:
    count+=1
    if not w in uniFreq:
      uniFreq[w] = 1 
    else:
      uniFreq[w] += 1 
  return uniFreq,count

```

```python
unigrams,unigramsCount = unigramFreq(DataCleaned)
unigrams= sorted(unigrams.items(), key=lambda x: x[1], reverse=True) 
```

```python
uniProbabilities = {}
log_probs={}
for word in unigrams:
  uniProbabilities[word[0]] = word[1] / unigramsCount
  log_probs[word[0]] = math.log(word[1] / unigramsCount)
```

```python
def makingBigrams(corpus):
  bigrams = list()
  for word in range(len(corpus)-1):
    w1 = corpus[word]
    w2 = corpus[word+1]
    bigrams.append(tuple((w1,w2)))

  return bigrams
```

```python
def bigramFreq(corpus):
  bi_freq={}
  count_bi=0;
  for c in corpus:
    count_bi += 1
    if not c in bi_freq:
      bi_freq[c] = 1
    else:
      bi_freq[c] += 1 
      
  return bi_freq
```

```python
bigrams = makingBigrams(DataCleaned)
bigramFreqs = bigramFreq(bigrams)
```

```python
def BiProb(uniFreq, bi_freq,bigrams):
  bi_probs={}

  for b in bigrams:
    if not b in bi_probs:
      for fb in bi_freq:
        if ( b == fb ):
          bi_val = bi_freq[fb]
      for u in uniFreq:
        if b[0] == u[0]:
          uni_val = u[1]
              
      bi_probs[b] = bi_val / uni_val
  return bi_probs
```

```python
BiProbablities =  BiProb(unigrams,bigramFreqs,bigrams)
print(BiProbablities)
```

```python
def makingTrigrams(corpus):
  trigrams = list()
  for word in range(len(corpus)-2):
    w1 = corpus[word]
    w2 = corpus[word+1]
    w3 = corpus[word+2]
    trigrams.append(tuple((w1,w2,w3)))

  return trigrams
```

```python
def trigramFreq(corpus):
  triFreq={}
  count=0;
  for word in corpus:
    count = count+1
    if not word in triFreq:
      triFreq[word] = 1 
    else:
      triFreq[word] += 1

  return triFreq
```

```python
trigrams = makingTrigrams(DataCleaned)
trigramsFreqs = trigramFreq(trigrams)
```

```python
def TriProb(bifreq, triFreq, trigrams):
  triProbabilites={}
  for word in trigrams:
    if not word in triProbabilites:
      for tf in triFreq:
        if ( word == tf ):
          triVal = triFreq[tf]
      for words in bifreq:
        if words[0] == word[0] and words[1] == word[1]:
          biVal = bifreq[words]
              
      triProbabilites[word] = triVal / biVal
  return triProbabilites

```

```python
TriPprobabilities =  TriProb(bigramFreqs,trigramsFreqs,trigrams)
```

```python
def genUniSentence(corpus):
  avgLength = 7
  sentences = list()
  for x in range(3):
    length = random.randint(avgLength,avgLength+2)
    flag = True
    start = ListOfStartWords[random.randint(0,len(ListOfStartWords))]
    end = ListOfEndWords[random.randint(0,len(ListOfEndWords))]

    sentence = start
    for y in range(1,length-1): 

      flag = True
      mid_word_index =  random.randint(0,len(corpus))
      rWord = corpus[mid_word_index][0]
      while(flag == True):
        if not rWord in sentence:
          sentence = sentence + ' '+ rWord
          flag = False
        else:
          flag = True  
      
    sentence = sentence + ' ' + end
    sentences.append(sentence)
  return sentences
```

```python
def genBiSentence(unigrams, bigramsProbs):
  avgLength = 7
  length = random.randint(avgLength,avgLength+2)
  sentence = []

  start = random.choice(unigrams)
  rWord = start[0]

  max = -1
  generatedBigram = ""

  for x in bigramsProbs:
    if x[0] == rWord:
      if bigramsProbs[x] > max:
        max = bigramsProbs[x]
        generatedBigram = x

  sentence.append(generatedBigram[0])
  sentence.append(generatedBigram[1])

  index = 1
  for x in range(0,length-1):
    generatedBigram = ""
    max = -1
    for y in bigramsProbs:      
      if y[0] == sentence[index] and y[1] not in sentence: 
        if (bigramsProbs[y] > max):
          max = bigramsProbs[y]
          generatedBigram = y
    
    sentence.append(generatedBigram[1])
    index = index + 1
  
  final_sentence = ""
  for x in range(0,len(sentence)):
    final_sentence = final_sentence + sentence[x] + ' '
  final_sentence = final_sentence[:-1]
  return final_sentence
```

```python
def genTriSentence(unigrams, bigramsProbs, trigramProbs):
  avgLength = 7
  length = random.randint(avgLength,avgLength+2)
  sentence = []

  start = random.choice(unigrams)
  rWord = start[0]

  bigram_max = -1
  generatedBigram = ""
  generatedTrigram = ""
  trigram_max = -1
  for x in bigramsProbs:
    if x[0] == rWord:
      if bigramsProbs[x] > bigram_max:
        bigram_max = bigramsProbs[x]
        generatedBigram = x
  for y in trigramProbs:
    if (y[0] == generatedBigram[0]) and (y[1]==generatedBigram[1]):
      if trigramProbs[y] > trigram_max:
        trigram_max = trigramProbs[y]
        generatedTrigram = y
  
  sentence.append(generatedTrigram[0])
  sentence.append(generatedTrigram[1])


  index = 1

  for i in range(0, length-5):
    generatedBigram = ""
    bigram_max = -1
    for b in bigramsProbs:
      if b[0] == sentence[index]:
        bigram_max = bigramsProbs[b]
        generatedBigram = b
    
    generatedTrigram = ""
    trigram_max = -1

    for t in trigramProbs:
      if (t[0] == generatedBigram[0]) and (t[1] == generatedBigram[1]):
        if (trigramProbs[t] > trigram_max):
          trigram_max = trigramProbs[t]
          generatedTrigram = t
    
    sentence.append(generatedTrigram[1])
    sentence.append(generatedTrigram[2])
    index = index + 2

  final_sentence = ""
  for i in range(0,len(sentence)):
    final_sentence = final_sentence + sentence[i] + ' '
  final_sentence = final_sentence[:-1]

  return final_sentence
```

```python
def Laplace(grams, probs, sentence):
  for i in grams:
    if i not in probs:
      probs[i] = 0
  
  for j in probs:
    v = probs[j] 
    n = v + 1 
    d = len(sentence) + len(probs)
    new_v = n / d
    probs[j]= new_v
```

```python
def uniPP(unigram_sentence, unigrams_probs):
  splitted = unigram_sentence.split()
  sum=1
  for i in splitted:
    if i not in uniProbabilities:
      Laplace(splitted,unigrams_probs,unigram_sentence)
  
  for i in splitted:
    sum = sum + math.log(uniProbabilities[i]) 
  N = len(splitted)
  perplexity =  (sum) * -1/N
  return perplexity
  
```

```python
def BiPP(biSentence, biProbs):
  bigrams_sentence = makingBigrams(biSentence.split())
  
  product=1

  for i in bigrams_sentence:
    if i not in biProbs:
      Laplace(bigrams_sentence,biProbs,biSentence)


  for i in bigrams_sentence:
    product = product + math.log(biProbs[i])

  N = len(bigrams_sentence)
  perplexity =  (product) * -1/N
  return perplexity
```

```python
def BisPP(BiSentences, biProbs):
  totalPP = 1
  p=list()
  for i in range(0,len(BiSentences)):
    v = BiPP(BiSentences[i], biProbs)
    v = math.log(v,2)
    p.append(v)

  for i in range(0,len(p)):
    totalPP = p[i]+ totalPP

  return totalPP
```

```python
def TriPP(triSentence, triProbs):
  trigrams_sentence = makingTrigrams(triSentence.split())
  product=1


  for i in trigrams_sentence:
    if i not in triProbs:
      Laplace(trigrams_sentence,triProbs,triSentence)

  for i in trigrams_sentence:
    product = product + math.log(triProbs[i])
  

  N = len(trigrams_sentence)
  perplexity =  (product) * -1/N
  return perplexity

```

```python
def TrisPP(TriSentences, triProbs):
  totalPP = 0
  p=list()
  for i in range(0,len(TriSentences)):
    v = TriPP(TriSentences[i], triProbs)
    v = math.log(v,2)
    p.append(v)

  for i in range(0,len(p)):
    totalPP = p[i]+ totalPP

  return totalPP
```
